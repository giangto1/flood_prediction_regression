# Feature Selection: PCA

```python
X = train.iloc[:,0:19].values
y = train.iloc[:, 20].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 19, svd_solver="full")

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)

explained_variance = pca.explained_variance_ratio_
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.show()
```

```python
print ("Proportion of Variance Explained : ", pca.explained_variance_ratio_)  
    
out_sum = np.cumsum(pca.explained_variance_ratio_)  
print ("Cumulative Prop. Variance Explained: ", out_sum)

# Proportion_of_Variance_Explained :  [0.32272604 0.15399155 0.12394602 0.10037712 0.06864445 0.05877295]
# Cumulative_Prop_Variance_Explained:  [0.32272604 0.47671759 0.60066362 0.70104074 0.76968519 0.82845815]
```
